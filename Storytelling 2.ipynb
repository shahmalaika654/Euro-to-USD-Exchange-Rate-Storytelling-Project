## Libraries Import and Reading the Dataset

**In the initial code cells, we begin by importing the necessary libraries, reading a CSV file containing exchange rate data, and performing data cleaning and preprocessing**

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
exchange_rate = pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
exchange_rate
```

## Data Pre-processing

```python
exchange_rate.info()
```

## Exploratory Data Analysis

**We rename columns, convert date values to datetime objects, sort the data by time, and remove rows with missing or invalid data.**

```python
exchange_rate.rename(columns = {"[US dollar ]":"US_dollar","Period\\Unit:":"Time"},inplace = True)
```

```python
exchange_rate["Time"] = pd.to_datetime(exchange_rate["Time"])
```

```python
exchange_rate.sort_values("Time", inplace = True)
```

```python
exchange_rate.reset_index(drop = True, inplace = True)
```

```python
exchange_rate.head()
```

```python
euro_to_dollar = exchange_rate[["Time","US_dollar"]].copy()
euro_to_dollar["US_dollar"].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar.US_dollar != "-"]
```

```python
euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)
```

```python
euro_to_dollar["US_dollar"].rolling(7).mean().head(10)
```

## Visualizaing Exchange Rate Data

**In this section, we create line plots to visualize the exchange rate data. We first create a basic line plot of the exchange rate over time. Then, we explore the data by calculating rolling means with different window sizes and create multiple subplots to compare them. The subplots provide insights into the data's trends and fluctuations.**

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)

plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.title("Orginal Values", weight = "bold")


for i, rolling_mean in zip(range(2,7), [7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar["Time"], 
             euro_to_dollar["US_dollar"].rolling(rolling_mean).mean())
    plt.title("Rolling Window:" + str(rolling_mean), weight = "bold")

plt.tight_layout()
```

```python
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
```

## Analyzing the Financial Crisis

**In this section, we focus on a specific time period: the financial crisis. We filter the data to select the relevant years and calculate the maximum exchange rate during this period. We then prepare the plot to visualize this time frame and highlight the crisis period using different colors and line widths.**

```python
financial_crisis = \
euro_to_dollar[(euro_to_dollar["Time"].dt.year >= 2006) & (euro_to_dollar["Time"].dt.year <= 2009)]

financial_crisis_7_8 = \
euro_to_dollar[(euro_to_dollar["Time"].dt.year >= 2007) & (euro_to_dollar["Time"].dt.year <= 2008)]

financial_crisis_7_8["US_dollar"].max()
```

## Analyzing Different US Presidents' Terms
**We continue the story by dividing the data into segments representing the terms of different U.S. presidents (Bush, Obama, Trump). Each president's data is plotted with custom tick labels and additional text annotations to provide context and highlight key periods.**

```python
bush_obama_trump = \
euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar["Time"].dt.year < 2021)]
```

```python
bush = bush_obama_trump[bush_obama_trump.Time.dt.year < 2009]
obama = bush_obama_trump[(bush_obama_trump.Time.dt.year >= 2009) & \
                         (bush_obama_trump.Time.dt.year < 2017)]
trump = bush_obama_trump[(bush_obama_trump.Time.dt.year >= 2017) & \
                         (bush_obama_trump.Time.dt.year < 2021)]
```

## Creating a Composite Plot
**In this cell, we create a composite plot that overlays the exchange rate data for all three presidents, making it easy to compare their terms. We set custom tick labels for the x-axis and add informative text annotations to the plot. This composite plot is a critical part of the storytelling, as it shows the entire narrative in a single view.`**

```python
plt.figure(figsize=(15,5))

ax1 = plt.subplot(2,3,1)
ax1.plot(bush["Time"], bush["rolling_mean"], linewidth = 3, color = "#bd55fb")
ax1.set_xticklabels(['',"2001","","2003","","2005","","2007","",'2009'], alpha = 0.3, fontdict={"fontsize":11})
ax1.set_yticks([1.0,1.2,1.4,1.6])
ax1.set_yticklabels(["1.0","1.2","1.4","1.6 $"], alpha = 0.3, fontdict={"fontsize":11})
ax1.text(12500.0, 1.73, "BUSH",color='#bd55fb', weight="bold")
ax1.text(12300.0, 1.65, "(2001-2009)", alpha=0.5, fontdict={"fontsize": 11})



ax2 = plt.subplot(2,3,2)
ax2.plot(obama["Time"], obama["rolling_mean"], linewidth = 3, color = "#fba40e")
ax2.set_xticklabels(['',"2009","","2011","","2013","","2015","",'2017'], alpha = 0.3, fontdict={"fontsize":11})
ax2.set_yticks([0.8,1.0,1.2,1.4,1.6])
ax2.set_yticklabels(['',"1.0","1.2","1.4","1.6 $"], alpha = 0.3, fontdict={"fontsize":11})
ax2.text(15300.0, 1.73, "OBAMA",color='#bd55fb', weight="bold")
ax2.text(15250.0, 1.65, "(2009-2017)", alpha=0.5, fontdict={"fontsize": 11})

ax3 = plt.subplot(2,3,3)
ax2 = plt.subplot(2,3,2)
ax3.plot(trump["Time"], trump["rolling_mean"], linewidth = 3, color = "#12b5ec")
ax3.set_xticklabels(['',"2017","","2018","","2019","","2020","",'2021'], alpha = 0.3, fontdict={"fontsize":11})
ax3.set_yticks([0.8,1.0,1.2,1.4,1.6])
ax3.set_yticklabels(['',"1.0","1.2","1.4","1.6 $"], alpha = 0.3, fontdict={"fontsize":11})
ax3.text(17700.0, 1.73, "TRUMP",color='#bd55fb', weight="bold")
ax3.text(17650.0, 1.65, "(2017-2021)", alpha=0.5, fontdict={"fontsize": 11})



ax4 = plt.subplot(2,1,2)
ax4.plot(bush_obama_trump["Time"], bush_obama_trump["rolling_mean"], linewidth = 1.5, color = "#A6D785")
ax4.plot(bush["Time"], bush["rolling_mean"], linewidth = 3, color = "#bd55fb")
ax4.plot(obama["Time"], obama["rolling_mean"], linewidth = 3, color = "#fba40e")
ax4.plot(trump["Time"], trump["rolling_mean"], linewidth = 3, color = "#12b5ec")


ax4.set_xticks([])
ax4.set_yticks([1.0,1.2,1.4,1.6])
ax4.set_yticklabels(["1.0","1.2","1.4","1.6 $"], alpha = 0.3, fontdict={"fontsize":11})

ax4.text(10500.0, 2.9, "Euro-USD rate averaged 1.22 under the last three US presidents", weight="bold", fontdict={"fontsize": 18})
ax4.text(10500.0, 2.8, "Euro-USD exchange rate under George W. Bush (2001-2009), Barack Obama (2009-2017) and Donald Trump (2017-2021)")
ax4.text(10500.0, 0.67, "(C) Yuaan Hussain"+ " "*130 + "Source: European Central Bank",
        color = "#f0f0f0", backgroundcolor = "#4d4d4d", size = 15)

plt.show()
```
